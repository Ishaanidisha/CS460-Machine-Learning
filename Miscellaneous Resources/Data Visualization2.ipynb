```python
import matplotlib.pyplot as plt

# Example data
data = [1, 2, 2, 3, 3, 3, 4, 5, 5, 5, 5, 6]

# Create histogram
plt.hist(data, bins=5, edgecolor='black')

# Add titles and labels
plt.title('Histogram of Example Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
data = [1, 2, 2, 3, 3, 3, 4, 5, 5, 5, 5, 6]

# Create box plot
sns.boxplot(data=data)

# Add title and labels
plt.title('Box Plot of Example Data')
plt.ylabel('Values')

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
x = [1, 2, 3, 4, 5, 6]
y = [2, 3, 4, 5, 6, 7]

# Create scatter plot
sns.scatterplot(x=x, y=y)

# Add title and labels
plt.title('Scatter Plot of X vs Y')
plt.xlabel('X Variable')
plt.ylabel('Y Variable')

# Show plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example DataFrame (replace with your own data)
df = pd.DataFrame({
    'Variable1': [1, 2, 3, 4, 5],
    'Variable2': [5, 4, 3, 2, 1],
    'Variable3': [2, 3, 4, 5, 6],
    'Variable4': [7, 8, 7, 8, 7]
})

# Create pair plot
sns.pairplot(df)

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 35, 20, 20]
colors = ['gold', 'lightblue', 'lightgreen', 'lightcoral']

# Create pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Ensures pie is drawn as a circle

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
import numpy as np
data = np.random.rand(10, 10)  # Generating random data for the heat map

# Create heat map
sns.heatmap(data, annot=True, cmap='coolwarm')

plt.title('Heat Map Example')
plt.show()
```
