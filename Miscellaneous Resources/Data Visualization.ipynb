```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Data Generation# ---------------------------------------
np.random.seed(42)

# Simulate data for 200 customers and 500 transactions
customer_ids = np.random.randint(1000, 1100, size=500)
transaction_dates = pd.date_range(start='2024-01-01', periods=500, freq='D')
amount_spent = np.random.uniform(10, 500, size=500)
product_categories = np.random.choice(['Electronics', 'Clothing', 'Groceries', 'Furniture'], size=500)

data = {
    'Customer ID': customer_ids,
    'Transaction Date': np.random.choice(transaction_dates, size=500),
    'Amount Spent ($)': amount_spent,
    'Product Category': product_categories
}

df = pd.DataFrame(data)

# Mean, Median, and Standard Deviation for Amount Spent (Already Provided)
mean_spent = df['Amount Spent ($)'].mean()
median_spent = df['Amount Spent ($)'].median()
std_spent = df['Amount Spent ($)'].std()

print("Mean Amount Spent: $", round(mean_spent, 2))
print("Median Amount Spent: $", round(median_spent, 2))
print("Standard Deviation of Amount Spent: $", round(std_spent, 2))

# Frequency Distribution of Product Categories (Already Provided)
category_counts = df['Product Category'].value_counts()
print("\nFrequency Distribution of Product Categories:")
print(category_counts)

# Step 2: Additional Descriptive Statistics# ---------------------------------------
mode_spent = df['Amount Spent ($)'].mode()[0]  # Mode
variance_spent = df['Amount Spent ($)'].var()  # Varianceprint("Mode Amount Spent: $", round(mode_spent, 2))
print("Variance of Amount Spent: $", round(variance_spent, 2))

# Output for mode and varianceprint("\nCalculated Descriptive Statistics:")
print("Mode Amount Spent: $", round(mode_spent, 2))
print("Variance of Amount Spent: $", round(variance_spent, 2))

# Step 3: Data Visualization# ---------------------------------------

plt.figure(figsize=(12, 5))

# Bar Chart for Product Categories
plt.subplot(1, 2, 1)
sns.countplot(x='Product Category', data=df)
plt.title('Transaction Frequency by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Frequency')
plt.grid(True)

# Histogram for Amount Spent
plt.subplot(1, 2, 2)
sns.histplot(df['Amount Spent ($)'], bins=20, kde=True)
plt.title('Distribution of Amount Spent')
plt.xlabel('Amount Spent ($)')
plt.ylabel('Frequency')
plt.grid(True)

plt.tight_layout()
plt.show()

# Step 4: Advanced EDA# ---------------------------------------

plt.figure(figsize=(14, 6))

# Box Plot for Amount Spent by Product Category
plt.subplot(1, 2, 1)
sns.boxplot(x='Product Category', y='Amount Spent ($)', data=df)
plt.title('Amount Spent by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Amount Spent ($)')

# Correlation Matrix
plt.subplot(1, 2, 2)
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')

plt.tight_layout()
plt.show()
```
